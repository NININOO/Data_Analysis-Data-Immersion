```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn
corr = sns.heatmap(df_1.corr(), annot = True, ax = ax) 
```

```python
# Creating a categorical variable that splits the "Renewable energy share in the total final energy consumption (%)" column into categories

sns.histplot(df['Renewable energy share in the total final energy consumption (%)'], bins = 20, kde = True)
```
