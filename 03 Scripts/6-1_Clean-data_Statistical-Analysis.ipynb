### Clean Data

```python
#Check number of missing values
df.isna().sum()
```

```python
#Plot N/A variables
df.isna().sum().plot(kind = 'bar')
```

```python
#Finding duplicate rows
duplicate_rows = df[df.duplicated(keep='first')]
duplicate_rows.shape
```

```python
#Percentage of missing values in our dataset
(((df.isna().sum())/df.shape[0])*100).sort_values(ascending = False)
```

```python
#Drop columns with more than 25% missing values
df.drop(['Renewables (% equivalent primary energy)', 'Financial flows to developing countries (US $)', 'Renewable-electricity-generating-capacity-per-capita'], axis = 1, inplace = True)
```

```python
#Check again our dataset
df.info()
```

```python
df.columns
```

```python
# Identify numeric columns
numeric_columns = df.select_dtypes(include=['number'])

# Fill missing values with mean for numeric columns only
df[numeric_columns.columns] = numeric_columns.fillna(numeric_columns.mean())
```

```python
#fill remaining missing values with mean value of that column
df.fillna(df.mean(numeric_only=True).round(1), inplace=True)
```

```python
#Check again percentage of missing values in our dataset
(((df.isna().sum())/df.shape[0])*100).sort_values(ascending = False)
```

```python
#checking nulls again
df.isnull().sum()
```

```python
# find the first row with NaN values
row_with_nan = df[df.isna().any(axis=1)]

print(row_with_nan.iloc[0])
```

```python
# Define the condition to identify the row you want to drop
condition = (
    (df['Country'] == 'French Guiana') &
    (df['Year'] == 2000) &
    (df['Access to electricity (% of population)'] == 78.9)
)

# Drop the row that matches the condition
df.drop(df[condition].index, inplace=True)

# Reset the index if needed
df.reset_index(drop=True, inplace=True)
```
